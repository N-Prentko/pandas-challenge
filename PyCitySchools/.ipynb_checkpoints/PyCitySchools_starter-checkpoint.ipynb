```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

combined_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

total_number_schools = len(combined_data["school_name"].unique())
total_students = combined_data["Student ID"].count()

total_budget = school_data["budget"].sum()

avg_math_score = combined_data["math_score"].mean()
avg_reading_score = combined_data["reading_score"].mean()
passing_math_count = combined_data[(combined_data["math_score"]>=70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(total_students)*100
passing_reading_count = combined_data[(combined_data["reading_score"]>=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(total_students)*100
overall_passing_parcentage = (avg_math_score + avg_reading_score)/2

print(total_number_schools,total_students,total_budget,avg_math_score,
     avg_reading_score,passing_math_count,passing_math_percentage,passing_reading_count,
     passing_reading_percentage,overall_passing_parcentage)


district_summary = pd.DataFrame({"Total Schools": [total_number_schools], 
                                 "Total Students": [total_students], 
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [avg_math_score], 
                                 "Average Reading Score": [avg_reading_score],
                                 "% Passing Math": [passing_math_percentage],
                                 "% Passing Reading": [passing_reading_percentage],
                                 "% Overall Passing": [overall_passing_parcentage]})


district_summary = district_summary[["Total Schools", "Total Students", "Total Budget",
                                     "Average Math Score","Average Reading Score",
                                     "% Passing Math","% Passing Reading",
                                     "% Overall Passing"]]

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)


district_summary
```

```python
#schooltype
school_type = school_data.set_index(["school_name"])["type"]
school_type
```

```python
#total students per school
total_students_per_school = combined_data["school_name"].value_counts()
total_students_per_school
```

```python
#total school budget
budget_per_school = combined_data.groupby(["school_name"]).mean()["budget"]
budget_per_school
```

```python
#budget per student 
budget_per_student = budget_per_school/total_students_per_school
budget_per_student
```

```python
#avg math scores per school
avg_math_scores_per_school = round(combined_data.groupby(["school_name"]).mean()["math_score"],2)
avg_math_scores_per_school
```

```python
#avergae reading scores per school
avg_reading_scores_per_school = round(combined_data.groupby(["school_name"]).mean()["reading_score"],2)
avg_reading_scores_per_school
```

```python
#students passing math by schools
#percentage passing math per schools
students_passing_math = combined_data[(combined_data["math_score"]>=70)]
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]
percentage_passing_math_schools = round((students_passing_math_per_school/total_students_per_school)*100,2)
percentage_passing_math_schools 
```

```python
#total students passing reading from the combined data
students_passing_reading = combined_data[combined_data["reading_score"]>=70]
#students_passing_reading

#find students passing reading per school
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]
#students_passing_reading_per_school

#percentage of students passing reading per school
percentage_passing_reading_per_school = round((students_passing_reading_per_school/total_students_per_school)*100,2)
percentage_passing_reading_per_school
```

```python
#overall passing rate per school
overall_passing_rate_per_school = (percentage_passing_math_schools + percentage_passing_reading_per_school)/2
overall_passing_rate_per_school
```

```python
#creating a summary of the school data
school_summary_data = pd.DataFrame({"School Type":school_type,
                                    "Total Students":total_students_per_school,
                                    "Total School Budget":budget_per_school,
                                    "Budget Per Student":budget_per_student,
                                    "Average Math Score Per School":avg_math_scores_per_school,
                                    "Average Reading Score Per School":avg_reading_scores_per_school,
                                    "% Passing Math Per School":percentage_passing_math_schools,
                                    "% Passing Reading Per School":percentage_passing_reading_per_school,
                                    "Overall Passes":overall_passing_rate_per_school})
school_summary_data
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

```python
#create bins for holding data
bins = [0, 600, 630, 660, 690]
spending_per_student = ["<=$600", "$601<=$630", "$631<=$660", "$661<=$690"]

school_summary_data["Spending/Student"] = pd.cut(budget_per_student,bins,labels = spending_per_student)
school_summary_data
```

## Scores by School Size

```python
#bins for schools based on size
bin_sizes = [0,1200,2400,3600,4800]
school_population = ["<=1200","1201<=2400","2401<=3600","3601<=4800"]

school_summary_data["School's Population"] = pd.cut(school_summary_data["Total Students"],bin_sizes,labels = school_population)
school_summary_data
```
